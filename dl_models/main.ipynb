```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8m.pt")

model.train(
    data="dataset/fire_dataset/data.yaml",

    # Balanced training time vs accuracy
    epochs=40,              # enough learning without wasting time
    imgsz=768,              # reduced slightly → faster + still good for small fire
    batch=6,                # safe for 6GB GPU

    device=0,
    amp=True,

    # Small fire optimization (important)
    scale=0.5,
    mosaic=1.0,
    mixup=0.1,              

    lr0=0.003,
    weight_decay=0.0005,

    patience=15,            # avoids overtraining
    close_mosaic=10,

    # ⚙️ Performance
    workers=4,
    cache=False
)
```
